```python
# Import necessary packages
import matplotlib.pyplot as plt # Overlay pandas and xarray plots
import rioxarray as rxr # Work with raster data
import xarray as xr
```

## Harmonize data

### The aspect and predicted precipitation plots for each grassland will be reprojected and matched to the soil pH DataArrays as the soil pH data has the highest resolution.

### Reproject and match all Sheyenne DataArrays to the Sheyenne Soil DataArray

```python
# Reproject and match sheyenne_aspect_da to the sheyenne_soil_ph_da
sheyenne_aspect_matched_da = (sheyenne_aspect_da
                              .rio.reproject_match(sheyenne_soil_ph_da))
sheyenne_aspect_matched_da.plot()

# Plot Sheyenne National Grasslands boundary on sheyenne_aspect_matched_da plot
sheyenne_grasslands_gdf.boundary.plot(ax = plt.gca(), color='black')

# I was unable to change the colorbar label to show units of aspect instead of meters.

plt.title('Sheyenne National Grasslands & Surrounding Area Aspect Reprojected & '
          'Matched to the Sheyenne Soil pH DataArray')
plt.show()
```

```python
# Reproject and match sheyenne_45_annual_precip_da to the sheyenne_soil_ph_da
sheyenne_45_annual_precip_matched_da = (sheyenne_45_annual_precip_da
                                        .rio.reproject_match(sheyenne_soil_ph_da))
sheyenne_45_annual_precip_matched_da.plot()

# Plot Sheyenne National Grasslands boundary on sheyenne_45_annual_precip_matched_da plot
sheyenne_grasslands_gdf.boundary.plot(ax = plt.gca(), color='black')

plt.title('Predicted Average Annual Precipitation (mm) from 2076-2080 '
          'for an RCP of 4.5 of the Sheyenne National Grasslands and '
          'Surrounding Areas, Reprojected & Matched to the Sheyenne Soil pH DataArray')
plt.show()
```

```python
# Reproject and match sheyenne_85_annual_precip_da to the sheyenne_soil_ph_da
sheyenne_85_annual_precip_matched_da = (sheyenne_85_annual_precip_da
                                        .rio.reproject_match(sheyenne_soil_ph_da))
sheyenne_85_annual_precip_matched_da.plot()

# Plot Sheyenne National Grasslands boundary on sheyenne_85_annual_precip_matched_da plot
sheyenne_grasslands_gdf.boundary.plot(ax = plt.gca(), color='black')

plt.title('Predicted Average Annual Precipitation (mm) from 2076-2080 '
          'for an RCP of 8.5 of the Sheyenne National Grasslands and '
          'Surrounding Areas, Reprojected & Matched to the Sheyenne Soil pH DataArray')
plt.show()
```
